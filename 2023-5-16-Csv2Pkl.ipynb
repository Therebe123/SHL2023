```python
import pandas as pd
import sys
import numpy as np
import os
import time
```

# 数据转化

```python
#定义加载csv格式文件的函数
def load_csv_all(path, chunksize=2000000):
    df_tmp = []
    count = 0
    mem_cost = 0
    chunks = pd.read_csv(path, index_col=0, chunksize=chunksize)
    for chunk in chunks:
        count += 1
        mem_cost += sys.getsizeof(chunk)
        df_tmp.append(chunk)
        del chunk
        if count % 10 == 0:
            print("the chunk %d has been stored..." % count)
            print("the mem-cost is now: %.3f MB \n" % (mem_cost / (1024 * 1024)))
    df = pd.concat(df_tmp)
    print('concat finished')
    print('n_chunks: %d' % count)
    print("the mem-cost is now: %.3f MB \n" % (sys.getsizeof(df) / (1024 * 1024)))
    return df
```

```python
#列出想要转化的文件的名字列表
file_dir="../SHL2023/data/"
all_file_list=os.listdir(file_dir)
all_csv_list=all_file_list[1:]
all_csv_list
```

```python
print(len(all_csv_list))
```

```python
#进行转化
for csv in all_csv_list:
    start=time.time()
    print('------------------------ Loading {} ------------------------'.format(csv))
    csv_path=os.path.join(file_dir,csv)
    df=load_csv_all(csv_path)
    df.to_pickle("../SHL2023/data_pkl/"+csv)
    end=time.time()
    print("数据转化完成，用时 {}s".format(end-start))
```

由于服务器存储空间不足，在转化csv文件为pkl文件后我们删除了csv文件，如果需要csv格式文件请重新运行Txt2Csv.ipynb文件
